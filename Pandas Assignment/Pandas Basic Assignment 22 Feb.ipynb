Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
import pandas as pd

# Create multiple Series
names = pd.Series(["Amit", "Ravi", "Neha"], index=[0, 1, 2])
ages = pd.Series([22, 25, 23], index=[0, 1, 2])
marks = pd.Series([78, 85, 90], index=[0, 1, 2])

# Create DataFrame using Series
df = pd.DataFrame({
    "Name": names,
    "Age": ages,
    "Marks": marks
})

print(df)
```
